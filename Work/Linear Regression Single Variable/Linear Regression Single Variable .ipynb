```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('canada_per_capita_income.csv')

df.rename(columns={'per capita income (US$)': 'percapitaincome'}, inplace=True)
```

```python
%matplotlib inline
plt.scatter(df.year,df.percapitaincome,color='blue',marker='.')

plt.xlabel('year')
plt.ylabel('Per Capita Income')
```

```python
reg=linear_model.LinearRegression()
reg.fit(df[['year']].values,df['percapitaincome'])
reg.get_params()
```

```python
p = reg.predict([[2020]])
```

```python
%matplotlib inline

plt.xlabel('year')
plt.ylabel('Per Capita Income')

plt.scatter(df.year,df.percapitaincome,color='blue',marker='.')
plt.plot(df.year,reg.predict(df[['year']].values),color='red')
```
